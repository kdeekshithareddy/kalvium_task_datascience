# Beautifull soup used


````python
url = "https://results.eci.gov.in/PcResultGenJune2024/index.htm"
req = requests.get(url)
````

````python
req
````

````python
soup = BeautifulSoup(req.content)
````

````python
num = []
for i in soup.findAll('a'):
  num.append(i)
````

````python
num_req = num[7:49]
````

````python
total_seats = []
for i in range(len(num_req)):
  i = re.sub('<[^>]*>', '', str(num_req[i])) # Convert Tag to string before applying rege
  total_seats.append(i)
````

````python
total_seats
````

````python
name = []
for i in soup.findAll('td'):
  name.append(i)
all_name = name[::4]
````

````python

name_party = []
for i in range(len(all_name)):
  i = re.sub('<[^>]*>', '', str(all_name[i])) # Convert Tag to string before applying rege
  name_party.append(i)
````

````python
name_party
````

````python


name_party = name_party[:len(total_seats)]

df = pd.DataFrame({
    "Name Party": name_party,
    "Total Seats": total_seats,

})
````

````python
df.head()
````

````python
# @title Distribution of Total Seats

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['Total Seats'], bins=10, edgecolor='black')
plt.xlabel('Total Seats')
plt.ylabel('Frequency')
_ = plt.title('Distribution of Total Seats')
````

````python
# @title Top 10 Parties by Total Seats

import matplotlib.pyplot as plt

df_top10 = df.sort_values('Total Seats', ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.bar(df_top10['Name Party'], df_top10['Total Seats'])
plt.xlabel('Name Party')
plt.ylabel('Total Seats')
plt.xticks(rotation=90)
_ = plt.title('Top 10 Parties by Total Seats')
````

````python
# @title Seat Distribution Among Political Parties

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.pie(df['Total Seats'], labels=df['Name Party'], autopct='%1.1f%%', startangle=90)
_ = plt.title('Seat Distribution Among Political Parties')
````

````python
url1 = "https://results.eci.gov.in/AcResultByeJune2024/"
req = requests.get(url1)
````

````python
soup = BeautifulSoup(req.content)
````

````python
soup
````

````python
name_n = []
for i in soup.findAll('h5'):
  name_n.append(i)
````

````python

name_bye = []
for i in range(len(name_n)):
  i = re.sub('<[^>]*>', '', str(name_n[i])) # Convert Tag to string before applying rege
  name_bye.append(i)
````

````python
name_bye
````

````python
len(name_bye)
````

````python
state = []
for i in soup.findAll('h4'):
  state.append(i)
````

````python

state_bye = []
for i in range(len(state)):
  i = re.sub('<[^>]*>', '', str(state[i])) # Convert Tag to string before applying rege
  state_bye.append(i)
````

````python
state_bye
````

````python
len(state_bye)
````

````python
party_n = []
for i in soup.findAll('h6'):
  party_n.append(i)
````

````python
party_n = party_n[1:26]
````

````python
party_n
````

````python
len(party_n)
````

````python
party_bye = []
for i in range(len(party_n)):
  i = re.sub('<[^>]*>', '', str(party_n[i])) # Convert Tag to string before applying rege
  party_bye.append(i)
````

````python
party_bye
````

````python
len(party_bye)
````

````python


# Adjust the length of the longest list to match the shortest
min_length = max(len(name_bye), len(party_bye), len(state_bye))
name_bye = name_bye[:min_length]
party_bye = party_bye[:min_length]
state_bye = state_bye[:min_length]

df = pd.DataFrame({
    "Name Bye": name_bye,
    "Party Bye": party_bye,
    "State Bye": state_bye
})
````

````python
df
````

````python
# @title Distribution of Representatives

import matplotlib.pyplot as plt
import seaborn as sns

df_heatmap = df.pivot_table(index='State Bye', columns='Party Bye', values='Name Bye', aggfunc='count', fill_value=0)

plt.figure(figsize=(12, 8))
sns.heatmap(df_heatmap, annot=True, cmap='viridis', fmt='g')
plt.xlabel('Party')
plt.ylabel('State')
_ = plt.title('Distribution of Representatives')
````

````python
# @title Distribution of Representatives Across States

import matplotlib.pyplot as plt

df['State Bye'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Representatives Across States')
_ = plt.ylabel('')
````

````python
# @title State Diversity of Party Representation

import matplotlib.pyplot as plt

df.groupby('Party Bye')['State Bye'].nunique().sort_values().plot(kind='barh')
plt.title('Number of States Represented by Each Party')
plt.xlabel('Number of States')
_ = plt.ylabel('Party')
````

````python
# @title Party Representation by State

import matplotlib.pyplot as plt
df.groupby(['State Bye', 'Party Bye'])['Name Bye'].count().unstack().plot(kind='bar', stacked=True)
plt.xlabel('State')
plt.ylabel('Number of Representatives')
_ = plt.title('Party Representation by State')
````
